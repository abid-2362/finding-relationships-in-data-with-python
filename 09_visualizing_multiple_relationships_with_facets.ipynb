```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
automobile_data = pd.read_csv('data/automobiles_data_processed.csv')
automobile_data.head()
```

```python
automobile_data.dtypes
```

```python
plt.figure(figsize=(15, 10))

fg = sns.FacetGrid(automobile_data,
                   col='fuel-type',
                   height=7, aspect=1)

fg.map(sns.histplot, 'price', color='red', kde=True)
plt.show()
```

```python
tips_data = sns.load_dataset('tips')
tips_data.head()
```

```python
plt.figure(figsize=(15, 10))

fg = sns.FacetGrid(tips_data,
                   col='size',
                   height=7, aspect=1)

fg.map(sns.histplot, 'total_bill', color='green', kde=True)
plt.show()
```

```python
np.sort(tips_data['size'].unique())
```

```python
group_size_values = np.sort(tips_data['size'].unique())
plt.figure(figsize=(15, 10))

fg = sns.FacetGrid(tips_data,
                   # row='time',
                   col='time',
                   height=6, aspect=1)

fg.map(sns.swarmplot, 'size', 'total_bill', order=group_size_values)
plt.show()
```
