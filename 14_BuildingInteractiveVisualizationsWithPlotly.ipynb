```python
import plotly
import pandas as pd
import plotly.graph_objs as go
import plotly.offline as offline

offline.init_notebook_mode(connected=True)
```

```python
plotly.__version__
```

```python
nifty_data = pd.read_csv('data/NSEI.csv')
nifty_data.head()
```

```python
nifty_data.shape
```

```python
nifty_data.dtypes
```

```python
nifty_data['Date'] = pd.to_datetime(nifty_data['Date'])
```

```python
data_2019 = nifty_data.loc[nifty_data['Date'].dt.year == 2019]
```

```python
data_2019.to_csv('data/NSEI_2019.csv')
```

```python
nifty_data = data_2019.copy()
```

```python
nifty_data.shape
```

```python
nifty_data.isnull().sum()
```

```python
nifty_data.reset_index(inplace=True, drop=True)
```

```python
nifty_data.head()
```

```python
nifty_data.drop(columns=['Adj Close', 'Volume'], inplace=True)
```

```python
nifty_data.isnull().sum()
```

```python
trace = go.Scatter(
    x= nifty_data['Date'],
    y=nifty_data['Close'],
    mode='markers',
    marker=dict(
        size=7,
        color=nifty_data['Close'],
        colorscale='Rainbow',
        showscale=True,
        opacity=0.5
    ),
    text=nifty_data['Close']
)

data = [trace]

layout = go.Layout(
    title='Stocks',
    hovermode='closest',
    xaxis=dict(title='Date'),
    yaxis=dict(title='Close')
)

figure = go.Figure(data=data, layout=layout)
offline.iplot(figure)
```

```python
nifty_data['Month'] = nifty_data['Date'].dt.month
nifty_data.sample(5)
```

```python
nifty_data['Month'] = pd.Categorical(nifty_data['Month'])
```

```python
trace_box = go.Box(
    x = nifty_data['Month'],
    y = nifty_data['Close']
)
data = [trace_box]
offline.iplot(data)
```

```python
trace = go.Candlestick(
    x=nifty_data.Date,
    open=nifty_data.Open,
    high=nifty_data.High,
    low=nifty_data.Low,
    close=nifty_data.Close
)

data = [trace]

layout = go.Layout(
    title= 'Stocks',
    hovermode='closest',
    xaxis=dict(title='Date'),
    yaxis=dict(title='Close')
)

fig = go.Figure(data=data, layout=layout)
offline.iplot(fig)
```
