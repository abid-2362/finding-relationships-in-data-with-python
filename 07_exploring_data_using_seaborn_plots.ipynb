```python
import datetime
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_data = sns.load_dataset('tips')
tips_data.sample(10)
```

```python
tips_data.shape
```

```python
tips_data.describe()
```

```python
tips_data.groupby(by='time').mean()
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x='time', y='total_bill', data=tips_data)

plt.title('Bill and Tips')
plt.xticks(rotation=90)

plt.show()
```

```python
tip_mean = tips_data['tip'].mean()
tip_mean
```

```python
tips_data['above average'] = (tips_data['tip'] - tip_mean) > 0
tips_data[['tip', 'above average']].head()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(x='time',
              hue='above average',
              data=tips_data,
              order=tips_data['time'].value_counts().index
              )

plt.title('Bill and Tips')
plt.show()
```

```python
tips_data['time'].value_counts()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(x='time',
              hue='sex',
              data=tips_data,
              order=tips_data['time'].value_counts().index
              )

plt.title('Bill and Tips')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.boxplot(x='sex',
            y='total_bill',
            palette='nipy_spectral',
            data=tips_data
            )

plt.title('Bill and Tips')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.boxplot(x='size',
            y='total_bill',
            palette='nipy_spectral',
            data=tips_data
            )

plt.title('Bill and Tips')
plt.show()
```

```python
sns.catplot(x='size',
            y='total_bill',
            hue='time',
            data=tips_data,
            kind='bar', height=6, aspect=2,
            palette='CMRmap'
            )
plt.title('Bill and Tips')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.stripplot(x='size', y='tip', data=tips_data)

plt.title('Bill and Tips')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.swarmplot(x='size', y='tip', data=tips_data)

plt.title('Bill and Tips')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.swarmplot(x='size', y='tip', hue='above average', data=tips_data)

plt.title('Bill and Tips')
plt.show()
```

```python
plt.figure(figsize=(20, 8))

sns.swarmplot(x='size', y='tip', hue='above average', data=tips_data, dodge=True)

plt.title('Bill and Tips')
plt.show()
```
